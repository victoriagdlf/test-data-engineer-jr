De lo obtenido anteriormente se requiere relacionar las listas de vuelos y pasajeros, que permitan analizar el perfil del pasajero por cada vuelo efectuado. De tal forma que se puedan obtener datos consolidados. En esta parte deberás:

* Explicar el proceso que utilizado para unir los pasajeros y los vuelos: 
Cada uno de los vuelos tiene un cliente asignado, en este caso se asume que la clave del cliente y el ID del pasajero son lo mismo. Esto nos permite asigar a cada vuelo el nombre y la edad del pasajero en cuestión. 
* Qué tipo de relación y por qué.
Se eligió que el DataFrame base fuese el de vuelos debido a que varios vuelos pueden tener un mismo pasajero. Si se asignan los valores de cada vuelo usado por cada pasajero, la calidad de los datos bajaría pues no se tendría de forma tan clara los resultados ni su acomodo. 

```python
# Para cada vuelo asignar el pasajero correspondiente
Vuelos1617= Vuelos1617.rename(columns={'Cve_Cliente': 'ID_Pasajero'})

PV1617 = pd.merge(Vuelos1617, Pasajeros1617, on='ID_Pasajero')
PV1617
```

## Sección 5

Por último, se requiere el promedio semestral (el primer semestre es de Ene - Jun y el segundo es de Jul - Dic) del precio agrupado por Año, Clase, Ruta y las Línea Aérea como columnas.

```python
# Convertir la columna de fecha al tipo datetime
PV1617['Viaje'] = pd.to_datetime(PV1617['Viaje'], format='%m/%d/%Y')

# Agregar una columna de mes
PV1617['Mes'] = PV1617['Viaje'].dt.month

# Agregar una columna de año
PV1617['Año'] = PV1617['Viaje'].dt.year

PV1617['Semestre'] = PV1617['Mes'].apply(lambda x: 1 if x <= 6 else 2)

PV1617
```

```python
Avrg_price= PV1617.groupby(['Año','Semestre', 'Clase', 'Ruta', 'Linea_Aerea'])['Precio'].mean()


prev_max_rows = pd.get_option('display.max_rows')
pd.set_option('display.max_rows', None)

# Mostrar el DataFrame completo
Avrg_price

# Restaurar la configuración previa de visualización
pd.set_option('display.max_rows', prev_max_rows)
```
